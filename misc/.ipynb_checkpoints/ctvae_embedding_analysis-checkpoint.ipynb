```python
# first import embedding, standardization, compute a stacked bargraph showing the proportion of a vector's magnitude from 
import pandas as pd
from src.constants import Column
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
root = '/home/wangz222/scratch/embedding'
df = pd.read_pickle(f'{root}/ctvae-conditional_contrastive_model-44m9h8fu:v0.pkl')
```

```python
cell_count = df.groupby(Column.gene.value)[Column.gene.value].count()
```

```python
df_agg = df.groupby(Column.sgRNA.value).agg({col: 'mean' if col != Column.gene.value else 'first' for col in df.columns[1:]}
                                            ).groupby(Column.gene.value).mean()
df_standardized = df_agg - df_agg.loc['nontargeting']
```

```python
df_standardized.head()
```

```python
metadata = pd.read_pickle('/home/wangz222/scratch/splits/filtered_df.pkl')
def compute_proportion(series):
    counts = series.value_counts(normalize=True)
    return counts
mitotic_index = metadata.groupby(Column.gene.value)[Column.cell_cycle_stage.value].apply(compute_proportion).unstack().fillna(0)
ntc_mitotic_index = mitotic_index.loc['nontargeting', 'mitotic']
mitotic_index = (mitotic_index['mitotic']- ntc_mitotic_index)/ntc_mitotic_index * 100
```

```python
df_standardized['salient_mag'] = np.linalg.norm(df_standardized.iloc[:,:32], axis=1)
df_standardized['background_mag']  = np.linalg.norm(df_standardized.iloc[:,32:64], axis=1)
df_standardized['relative_contribution'] = df_standardized['background_mag']**2/(df_standardized['salient_mag']**2+df_standardized['background_mag']**2)
df_standardized.drop('nontargeting', inplace=True)

df_standardized['cell_count'] = cell_count
df_standardized['mitotic_index'] = mitotic_index
```

```python
df_standardized['salient_mag'] = df_standardized['salient_mag']/np.mean(df_standardized['salient_mag'])
df_standardized['background_mag'] = df_standardized['background_mag']/np.mean(df_standardized['background_mag'])
```

```python
plt.scatter(np.log10(df_standardized['salient_mag']+1), np.log10(df_standardized['background_mag']+1), s=1)
```

```python
df_standardized[df_standardized['salient_mag'] * 3 < df_standardized['background_mag']]
```

```python
top_mitotic = np.abs(df_standardized['mitotic_index'])>60
sub_mitotic = df_standardized.loc[top_mitotic, 'mitotic_index']
sub_vec_norm = df_standardized.loc[top_mitotic, 'relative_contribution']
```

```python
sub_vec_norm[(sub_vec_norm>0.85)*(sub_mitotic<200)].index
```

```python
sub_vec_norm[(sub_vec_norm<0.55)]
```

```python
plt.scatter(np.abs(sub_mitotic), sub_vec_norm)
plt.ylabel(r'$\frac{\|background\|^2}{\|background\|^2 + \|salient\|^2}$', fontsize=16)
plt.xlabel('% change in mitotic cells', fontsize=13)
# Compute Pearson correlation coefficient and the p-value
corr, p_value = pearsonr(np.abs(sub_mitotic), sub_vec_norm)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.text(500, 0.5, "Pearson: {:.3f} \n p-value: {:.3f}".format(corr, p_value), fontsize=12)
plt.savefig('scatterplot.svg', dpi=300)
```

```python
plt.scatter(df_standardized['cell_count'], df_standardized['relative_contribution'], 2)
plt.ylabel(r'$\frac{\|background embedding\|}{\|total embedding\|}$', fontsize=16)
plt.xlabel('Number of cells', fontsize=13)
# Compute Pearson correlation coefficient and the p-value
corr, p_value = pearsonr(df_standardized['cell_count'], df_standardized['relative_contribution'])
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.text(7700, 0.2, "Pearson: {:.3f} \n p-value: {:.3f}".format(corr, p_value), fontsize=12)
```
