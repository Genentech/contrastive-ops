```python
import pandas as pd
import numpy as np
import os
import sys
from joblib import Parallel, delayed
from tqdm import tqdm
import matplotlib.pyplot as plt

# Get all parquet files
DATA_DIR = '/projects/site/gred/resbioai/comp_vision/cellpaint-ai/ops/datasets/funk22/funk22-phenotype-profiles'
folder_path = f'{DATA_DIR}/mitotic-reclassified_cp_phenotype_normalized.parquet'
mitotic_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.parquet')]
folder_path = f'{DATA_DIR}/interphase-reclassified_cp_phenotype_normalized.parquet'
parquet_files = mitotic_files + [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.parquet')]
```

```python
def get_attr(file_path, attr):
    # open the Parquet file
    chunk = pd.read_parquet(file_path).reset_index()
    chunk = chunk.replace([np.inf, -np.inf], np.nan).dropna(how="any").reset_index(drop=True)

    if chunk.shape[0] > 0:
        return chunk[attr]
    
# Create the list of delayed tasks
attr = ['sgRNA_0', 'gene_symbol_0', 'plate','well','cell_gh2ax_mean', 'cell_gh2ax_median', 'cell_gh2ax_max', 'cell_gh2ax_upper_quartile', 'cell_dapi_int']
tasks = [delayed(get_attr)(file_path, attr) for file_path in parquet_files]
results = Parallel(n_jobs=-1, timeout=9999999)(tqdm(tasks))
aggregated = pd.concat(results)
```

```python
feature = attr[4:]
fig = plt.figure(figsize=(20,10))
gene = aggregated['gene_symbol_0'].unique()
plate = aggregated['plate'].unique()
well = aggregated['well'].unique()
aggregated['batch'] = aggregated['plate'] + aggregated['well']
batch = aggregated['batch'].unique()

for i, feat in enumerate(gene[:10]):
    ax = fig.add_subplot(4, 5, i+1)
    ax.hist(np.arcsinh(aggregated[(aggregated['gene_symbol_0']==feat) & (aggregated['cell_dapi_int'].between(0.7,1.2))]['cell_gh2ax_max']), alpha=0.2, density='probability', bins=200)
    ax.set_title(feat, fontsize=8)
    ax.set_xlabel('arcsinh(feature value)')
    ax.set_ylabel('proportion')
plt.subplots_adjust(hspace=0.8, wspace=0.5)
plt.show()
```
